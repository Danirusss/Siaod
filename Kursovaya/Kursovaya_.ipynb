```python
def bSearch(left,right,money,a):
    if (right == left):
        if (canBuy(left,money,a)!=-1):
            return (left,canBuy(left,money,a))
    if (right - 1 == left):
        if (canBuy(right,money,a)!=-1):
            return (right, canBuy(right,money,a))
        if (canBuy(left,money,a)!=-1):
            return (left,canBuy(left,money,a))
    mid = (right+left)//2
    if (canBuy(mid,money,a)!=-1):
        return bSearch(mid,right,money,a)
    else:
        return bSearch(left,mid-1,money,a)
    return -1;
def checkMas(arr,l, r):
    for i in range(len(arr)):
        if not(l <= arr[i] < r):
            print("ะะฒะตะดะตะฝะฝะพะต ะทะฝะฐัะตะฝะธะต ะฝะต ะฟัะธะฝะฐะดะปะตะถะธั ะทะฐะดะฐะฝะฝะพะผั ะพััะตะทะบั")
            return False
    return True

def checkVal(v,l, r):
    if l <= v < r:
        return True
    else:
        print("ะะฒะตะดะตะฝะฝะพะต ะทะฝะฐัะตะฝะธะต ะฝะต ะฟัะธะฝะฐะดะปะตะถะธั ะทะฐะดะฐะฝะฝะพะผั ะพััะตะทะบั")
        return False 
def canBuy(k,m,x):
    arr = x[:]
    s = 0
    for i in range(len(arr)):
        arr[i] += k*(i+1)
    arr.sort()
    s = sum(arr[:k])
    if (s<=m):
        return s
    return -1
try:
    while True:
        n, money = input('').split(' ')
        n, money = int(n) , int(money)
        if checkVal(n,1,10**5+1):
            if checkVal(money,1,10**9+1):
                break
    while True:
        arr = list(map(int, input().split()))
        if checkMas(arr,1,10**5+1):
            break
    a,b = bSearch(0,n,money,arr)
    print(a, b)
except Exception:
    print("ะะตะบะพััะตะบัะฝัะน ะฒะฒะพะด")

```

# 4 ะทะฐะดะฐะฝะธะต

```python
def boxed(a):
    a.sort()
    box=set()
    for x in a:
        if x-1>0 and x-1 not in box:
            box.add(x-1) #ะะพะฑะฐะฒะปัะตะผ ะฒ ัะปะพะฒะฐัั ะทะฝะฐัะตะฝะธะต, ะตัะปะธ ะฟะพะดัะพะดะธั ะฟะพ ััะปะพะฒะธัะผ ั ะพัะฝะธะผะฐะฝะธะตะผ ะพะดะฝะพะณะพ
        elif x not in box:
            box.add(x) #ะะพะฑะฐะฒะปัะตะผ ะฒ ัะปะพะฒะฐัั ะทะฝะฐัะตะฝะธะต, ะตัะปะธ ะฟะพะดัะพะดะธั ะฟะพ ััะปะพะฒะธัะผ ะฑะตะท ะพัะฝะธะผะฐะฝะธั
        elif x+1 not in box:
            box.add(x+1) #ะะพะฑะฐะฒะปัะตะผ ะฒ ัะปะพะฒะฐัั ะทะฝะฐัะตะฝะธะต, ะตัะปะธ ะฟะพะดัะพะดะธั ะฟะพ ััะปะพะฒะธัะผ ั ะดะพะฑะฐะฒะปะตะฝะธะตะผ
    return len(box)
try:
    while True:
        n = int(input())
        if checkVal(n,1,150000):
            break
    while True:
        arr = list(map(int, input().split()))
        if checkMas(arr,1,200000) & (len(arr) == n):
            break
    print(boxed(arr))
except Exception:
    print("ะะตะบะพััะตะบัะฝัะน ะฒะฒะพะด")
```

# 5 ะะฐะดะฐะฝะธะต

```python
from collections import deque
n, m = [int(x) for x in input().split()]
 
color = [0]*n
graph = [[] for i in range(0, n)]
for i in range(0, m):
    u, v = [int(x) - 1 for x in input().split()]
    graph[u].append(v)
    graph[v].append(u)

# ะะฒัะดะพะปัะฝัะน ะณัะฐั
res = True
for i in range(0, n):
    if color[i] != 0:
        continue
    q = deque()
    color[i] = 1
    q.append(i)
    while q:
        v = q.popleft()
        for a in graph[v]:
            if color[a] == 0:
                color[a] = -color[v]
                q.append(a)
            elif color[a] != -color[v]:
                res = False
                break
    if not res:
        break
if not res:
    print(-1)
else:
    s1 = []
    s2 = []
    for i in range(0, n):
        if color[i] == 1:
            s1.append(i)
        else:
            s2.append(i)
    print(len(s1))
    print(" ".join(str(x + 1) for x in s1))
    print(len(s2))
    print(" ".join(str(x + 1) for x in s2))
```

## 6 ะะฐะดะฐะฝะธะต

```python
n, m = map(int, input().split())
a = [[] for i in range(n)]
pos = [False] * n
for i in range(m):
    u, v = map(int, input().split())
    a[u - 1].append(v - 1)
    a[v - 1].append(u - 1)
otv = []
q = []
q.append(0)
while len(q) > 0:
    k = min(q)
    q.remove(k)
    pos[k] = True
    otv.append(k + 1)
    for i in a[k]:
        if not pos[i]:
            q.append(i)
            pos[i] = True
print(*otv)
```
