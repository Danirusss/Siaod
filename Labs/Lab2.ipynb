```python
#Метод Фибоначчи
import time
def FibonacciSearch(lys, val):
    fibM_minus_2 = 0
    fibM_minus_1 = 1
    fibM = fibM_minus_1 + fibM_minus_2
    while (fibM < len(lys)):
        fibM_minus_2 = fibM_minus_1
        fibM_minus_1 = fibM
        fibM = fibM_minus_1 + fibM_minus_2
    index = -1;
    while (fibM > 1):
        i = min(index + fibM_minus_2, (len(lys)-1))
        if (lys[i] < val):
            fibM = fibM_minus_1
            fibM_minus_1 = fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
            index = i
        elif (lys[i] > val):
            fibM = fibM_minus_2
            fibM_minus_1 = fibM_minus_1 - fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
        else :
            return i
    if(fibM_minus_1 and index < (len(lys)-1) and lys[index+1] == val):
        return index+1;
    return -1
def insert_vst(l, elem):
    index = FibonacciSearch(l,elem)
    l.insert(index, elem)
    return l
def delete_elem(l, elem):
    index = FibonacciSearch(l,elem)
    l.pop(index)
    return l
lys = [0,1,3,6,9]
val = 6
elem = 1
insert_vst(lys, elem)
print(lys)
delete_elem(lys, elem)
print(lys)
print(FibonacciSearch(lys, val))
```

```python
#Интерполяционный
def InterpolationSearch(lys, val):
    low = 0
    high = (len(lys) - 1)
    while low <= high and val >= lys[low] and val <= lys[high]:
        index = low + int(((float(high - low) / ( lys[high] - lys[low])) * ( val - lys[low])))
        if lys[index] == val:
            return index
        if lys[index] < val:
            low = index + 1;
        else:
            high = index - 1;
    return -1
def insert_vst(l, elem):
    index = InterpolationSearch(l,elem)
    l.insert(index, elem)
    return l
def delete_elem(l, elem):
    index = InterpolationSearch(l,elem)
    l.pop(index)
    return l
lys = [0,1,3,6,9]
val = 6
elem = 1
insert_vst(lys, elem)
print(lys)
delete_elem(lys, elem)
print(lys)
print(InterpolationSearch(lys, val))
```

```python
#Простое рехэширование
from random import randint
words = ['A', 'B', 'D', 'E', 'C']

def hashFunction(word):
    hash = 1
    for i in range(len(word)):
        hash *= (ord(word[i])+127)*341
    hash %= 1024
    return hash

class HashTable:
    def __init__(self):
        self.__size = 1024
        self.__table = []
        for i in range(self.__size):
            self.__table.append(None)

    def add(self, key, value):
        self.__table[key] = value

    def delete(self, key):
        self.__table[key] = None

    def contains(self, key):
        return self.__table[key] is not None

    def get(self, key):
        return self.__table[key]
    
    
class SimpleRehash:
    def __init__(self):
        self.table=HashTable()
        
    def add(self,value):
        hsh=hashFunction(value)
        if not self.table.contains(hsh):
            self.table.add(hsh,value)
        else:
            self.rehash(hsh,value)
            
    def rehash(self,hsh,value):
        while self.table.contains(hsh):
            hsh+=1
            if hsh>=1024:
                hsh %=1024
        self.table.add(hsh,value)
        
    def find(self,value):
        for i in range(1024):
            if self.table.get(i)==value:
                return i
        return -1
    
    def delete(self,value):
        hsh=self.find(value)
        if hsh>-1:
            self.table.delete(hsh)
        
    def print_table(self):
        for i in range(1024):
            if self.table.get(i) is not None:
                print(i,":",self.table.get(i))
                
arr=words.copy()
simple=SimpleRehash()

for i in range(len(arr)):
    simple.add(arr[i])
    
print("\nПростое рехеширование")
simple.print_table()
```
